```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
```

```python
A=1
B=3
omega=2*np.pi/10
N=100
```

```python
t=np.linspace(0,10,N)
#Entre punto y punto existe una cantidad específica
r=np.zeros((N,3))
```

```python
r[:,0]=A*np.cos(omega*t)
r[:,1]=A*np.sin(omega*t)
r[:,2]=B*t
```

```python
fig=plt.figure(figsize=(5,5))
ax=fig.add_subplot(111,projection='3d')

ax.scatter(r[:,0],r[:,1],r[:,2],color='k',marker='o',s=100)
ax.view_init()
```

```python
n=4
#Distretizar el eje x
x=np.linspace(0,1,n)
y=np.linspace(0,1,n)
X,Y=np.meshgrid(x,y)
#Convierte unidimensional a cuadricular
```

```python
r0=np.array([0.4,0.5])
```

```python
def GetE(r,r0,q=1,c=0):
    d=(r[0]-r0[0])**2 + (r[1]-r0[1])**2
    
    return q*(r[c]-r0[c])/d**(3/2)
    
```

```python
Ex=np.zeros((n,n))
Ey=np.zeros((n,n))

for i in range(n):
    for j in range(n):
        Ex[i,j]=GetE([x[1],y[j]],r0,1,0)
        Ey[i,j]=GetE([x[1],y[j]],r0,1,1)
```

```python
fig=plt.figure(figsize=(5,5))
ax=fig.add_subplot(111)
ax1=fig.add_subplot(122)
#ax.scatter(X,Y)

for i in range(n):
    for j in range(n):
        ax.quiver(x[i],y[j],Ex[i,j],Ey[i,j])

#Para pintar lineas de campo
ax1.streamplot(X,Y,Ex.T,Ey.T)
```
